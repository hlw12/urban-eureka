```python
import networkx as nx # 图数据控料片
import pandas as pd 
import numpy as np
import random #随机数
from tqdm import tqdm # 进度条
# 数据可视化
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv ("vocabulary.tsv", sep = "\t" )
```

```python
df.head()
```

```python
df.shape
```

```python
G = nx.from_pandas_edgelist(df, "source", "target", edge_attr=True,create_using=nx.Graph())
len(G)
```

```python
def get_randomwalk(node, path_length) :
    random_walk = [node]
    for i in range (path_length-1) :
        temp = list (G. neighbors (node) )
        temp = list (set (temp) - set (random_walk) )
        if len (temp) == 0:
            break
        random_node = random. choice (temp)
        random_walk. append (random_node)
        node = random_node
    return random_walk
```

```python
all_nodes = list(G.nodes())
```

```python
all_nodes
```

```python
get_randomwalk('earthquake engineering', 5)
```

```python
# 生成随机游走序列
gamma = 10
walk_length = 5
# 序列保存
random_walks = []
for n in tqdm(all_nodes):
    for i in range(gamma):
        random_walks.append(get_randomwalk(n, walk_length))
```

```python
len(random_walks)
```

```python
# word2vec
from gensim.models import Word2Vec#自然语言处理
```

```python
model = Word2Vec(vector_size=256,# Embedding維数
                  window=4,
                  sg=1, # Skip-Gram 1:中心词预测周围词/0:周围词预测中心词
                  hs=0, #不加分层softmax
                  alpha=0.03,
                  min_alpha=0.0007, #最小学习率
                  seed=14 # 随机数种子
                  )
```

```python
model.build_vocab(random_walks, progress_per=2) # 词汇表
```

```python
model.train(random_walks, total_examples=model.corpus_count, epochs=50, report_delay=1) # 训练 词->向量（低维，稠密）
```

```python
model.wv.similar_by_word('peak ground acceleration') # 探索输入的词在该表达空间中与哪一些词词意相近
```

```python
from sklearn. decomposition import PCA
# 可视化任务: 使用pageRank算法可视化前30个节点(PCA)
X = model.wv.vectors
pca = PCA(n_components=2)
embed_2d = pca.fit_transform(X)
pagerank = nx.pagerank(G)
node_importance = sorted(pagerank.items( ), key=lambda x:x[1], reverse=True)
```

```python
n = 30 # 30个节点
terms_chosen = []
for each in node_importance[:n]:
    terms_chosen.append(each [0])
```

```python
terms_chosen # terms_chosen.extend([.....])可以手动补充关心的节点
```

```python
term2index = model.wv.key_to_index 
```

```python
plt.figure(figsize=(14,14))
plt.scatter(embed_2d[:, 0], embed_2d[:, 1])

for item in terms_chosen:
    idx = term2index[item]
    plt.scatter(embed_2d[idx, 0], embed_2d[idx, 1], c='r', s=50)
    plt.annotate(item, xy=(embed_2d[idx, 0], embed_2d[idx, 1]), c='k', fontsize=12)
plt.show()
```

```python
# 使用TSNR降维算法（PCA-线性降维）
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, n_iter=1000)
embed_2d = tsne.fit_transform(X)
```

```python
plt.figure(figsize=(14,14))
plt.scatter(embed_2d[:, 0], embed_2d[:, 1])
plt.show()
```

```python
# 同上，对词条标记
plt.figure(figsize=(14,14))
plt.scatter(embed_2d[:, 0], embed_2d[:, 1])
for item in terms_chosen:
    idx = term2index[item]
    plt.scatter(embed_2d[idx, 0], embed_2d[idx, 1], c='r' ,s=50)
    plt.annotate(item, xy=(embed_2d[idx, 0], embed_2d[idx, 1]), c='k', fontsize=12)
plt.show()
```

```python
# TSNE 降维到2维后的embedding
terms_chosen_mask = np.zeros(X.shape[0])
for item in terms_chosen:
    idx = term2index[item]
    terms_chosen_mask[idx] = 1
```

```python
df = pd.DataFrame()
df['X'] = embed_2d[:, 0]
df['Y'] = embed_2d[:, 1]
df['item' ] = model.wv.index_to_key
df['pagerank'] = pagerank.values()
df['chosen'] = terms_chosen_mask
```

```python
df.to_csv('tsne_vis_2d.csv' ,index=False)
```
